```python
X = [50, 40, 60]
W = [2, 1, 3]
b = -200
y = 1
lr = 0.01
for epoch in range(20):
    z = sum(w*x for w, x in zip(W, X)) + b

    y_pred = 1 if z >= 0 else 0
    error = y - y_pred

    print(f"Epoch {epoch} | z={z:.2f} | pred={y_pred} | error={error}")

    if error == 0:
        break
    for i in range(len(W)):
        W[i] += lr * error * X[i]

    b += lr * error
print("\nFinal Weights:", W)
print("Final Bias:", b)


```
